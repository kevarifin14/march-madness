```python
def offensiveRatings(data):
    wOffRating = 100 * data.Wscore / (0.5 * \
    ((data.Wfga + 0.4 * data.Wfta - 1.07 * (data.Wor / (data.Wor + data.Ldr)) * \
    (data.Wfga - data.Wfgm) + data.Wto) + (data.Lfga + 0.4 * data.Lfta - 1.07 * \
    (data.Lor / (data.Lor + data.Wdr)) * (data.Lfga - data.Lfgm) + data.Lto)))   
    lOffRating = 100 * data.Lscore / (0.5 * \
    ((data.Lfga + 0.4 * data.Lfta - 1.07 * (data.Lor / (data.Lor + data.Wdr)) * \
    (data.Lfga - data.Lfgm) + data.Lto) + (data.Wfga + 0.4 * data.Wfta - 1.07 * \
    (data.Wor / (data.Wor + data.Ldr)) * (data.Wfga - data.Wfgm) + data.Wto)))   
    return wOffRating, lOffRating

def defensiveRatings(data):
    wDefRating = data.Lscore * 100 / (0.96 * (data.Lfga - data.Lor + data.Lto + (0.44 * data.Lfta)))
    lDefRating = data.Wscore * 100 / (0.96 * (data.Wfga - data.Wor + data.Wto + (0.44 * data.Wfta)))
    return wDefRating, lDefRating
    
def stats(df, teamId, season, daynum, games=8):
    try:
        wins = df.loc[season, :daynum - 1, teamId][WINNER_COLUMNS]
        losses = df.loc[season, :daynum - 1, :, teamId][LOSER_COLUMNS]
        wins["winPct"] = 1
        losses["winPct"] = 0
        winsStats = wins.rename(columns={WINNER_COLUMNS[i]: STAT_COLUMNS[i] for i in range(len(STAT_COLUMNS))})
        lossStats = losses.rename(columns={LOSER_COLUMNS[i]: STAT_COLUMNS[i] for i in range(len(STAT_COLUMNS))})

        selectedGames = pd.concat([winsStats, lossStats])[:games]
        return selectedGames[STAT_COLUMNS].mean()
    except: 
        return None

# print(stats(resultsDf, 1320, "2003-01-01", 69))
```

```python
teams = pd.DataFrame.from_csv("Teams.csv").to_dict()["Team_Name"]
tourneyResDf = pd.DataFrame.from_csv("TourneyDetailedResults.csv", index_col=["Season", "Daynum", "Wteam", "Lteam"])
regSeasonResDf = pd.DataFrame.from_csv("RegularSeasonDetailedResults.csv", index_col=["Season", "Daynum", "Wteam", "Lteam"])
resultsDf = pd.concat([regSeasonResDf, tourneyResDf])
resultsDf["Welo"] = 1600
resultsDf["Lelo"] = 1600

STAT_COLUMNS = ["score", "fgm", "fga", "fgm3", "fga3", "ftm", "fta", "or", "dr", "ast", "to", "stl", "blk", "pf", "elo"]
WINNER_COLUMNS = list(map(lambda c: "W" + c, STAT_COLUMNS)) 
LOSER_COLUMNS = list(map(lambda c: "L" + c, STAT_COLUMNS))

resultsDf["Worating"], resultsDf["Lorating"] = offensiveRatings(resultsDf[WINNER_COLUMNS + LOSER_COLUMNS])
resultsDf["Wdrating"], resultsDf["Ldrating"] = defensiveRatings(resultsDf[WINNER_COLUMNS + LOSER_COLUMNS])

resultsDf["Wfgp"] = resultsDf["Wfgm"] / resultsDf["Wfga"]
resultsDf["W3pp"] = resultsDf["Wfgm3"] / resultsDf["Wfga3"]
resultsDf["Wftp"] = resultsDf["Wftm"] / resultsDf["Wfta"]
resultsDf["Wftaperfga"] = resultsDf["Wfta"] / resultsDf["Wfga"]
resultsDf["Wtovp"] = 100 * resultsDf["Wto"] / (resultsDf["Wfga"] + 0.44 * resultsDf["Wfta"] + resultsDf["Wto"])
resultsDf["Wefgp"] = (resultsDf["Wfgm"] + 0.5 * resultsDf["Wfgm3"]) / resultsDf["Wfga"]

resultsDf["Lfgp"] = resultsDf["Lfgm"] / resultsDf["Lfga"]
resultsDf["L3pp"] = resultsDf["Lfgm3"] / resultsDf["Lfga3"]
resultsDf["Lftp"] = resultsDf["Lftm"] / resultsDf["Lfta"]
resultsDf["Lftaperfga"] = resultsDf["Lfta"] / resultsDf["Lfga"]
resultsDf["Ltovp"] = 100 * resultsDf["Lto"] / (resultsDf["Lfga"] + 0.44 * resultsDf["Lfta"] + resultsDf["Lto"])
resultsDf["Lefgp"] = (resultsDf["Lfgm"] + 0.5 * resultsDf["Lfgm3"]) / resultsDf["Lfga"]

WINNER_COLUMNS += ["Worating", "Wdrating", "Wfgp", "W3pp", "Wftp", "Wftaperfga", "Wtovp", "Wefgp"]
LOSER_COLUMNS += ["Lorating", "Ldrating", "Lfgp", "L3pp", "Lftp", "Lftaperfga", "Ltovp", "Lefgp"]
STAT_COLUMNS += ["orating", "drating", "fgp", "3pp", "ftp", "ftaperfga", "tovp", "efgp"]
resultsDf.sort_index(inplace=True)

resultsDf
```

```python
def updateElo(season, team, newElo):
    teamElos[season][team].append(newElo)

def getElo(season, team):
    return teamElos[season][team][-1]

def calcElo(winTeam, loseTeam, season):
    winnerRank = getElo(season, winTeam)
    loserRank = getElo(season, loseTeam)
    
    rankDiff = winnerRank - loserRank
    exp = (rankDiff * -1) / 400
    odds = 1 / (1 + math.pow(10, exp))
    if winnerRank < 2100: k = 32
    elif winnerRank >= 2100 and winnerRank < 2400: k = 24
    else: k = 16
    
    newWinnerRank = round(winnerRank + (k * (1 - odds)))
    newRankDiff = newWinnerRank - winnerRank
    newLoserRank = loserRank - newRankDiff
    
    return newWinnerRank, newLoserRank
```

```python
def initializeElos():
    return {season: {team: [1600] for team in teams.keys()} for season in range(2003, 2018)}

teamElos = initializeElos()

for i, (index, gameData) in enumerate(resultsDf.iterrows()):
    season, Daynum, Wteam, Lteam = index
    resultsDf.loc[index, "Lelo"] = getElo(season.year, Lteam)
    resultsDf.loc[index, "Welo"]  = getElo(season.year, Wteam)
    newWElo, newLElo = calcElo(Wteam, Lteam, season.year)
    updateElo(season.year, Lteam, newLElo)
    updateElo(season.year, Wteam, newWElo)

resultsDf
```

```python
NUM_FEATURES = 46

features = np.zeros(NUM_FEATURES)
labels = np.zeros(1)
```

```python
# left team beats right team - label is 1
# right team beats left team - label is 0

print(datetime.now())
for i, (index, g) in enumerate(resultsDf.iterrows()):
    if i % 1000 == 0: print("Iteration {0}".format(i))
    
    season, Daynum, Wteam, Lteam = index
    if Daynum < 60: continue
    seasonTs = "{0}-{1}-{2}".format(season.year, season.month, season.day)
    
    wStats = stats(resultsDf, Wteam, seasonTs, Daynum)
    lStats = stats(resultsDf, Lteam, seasonTs, Daynum)
    
    wStats["elo"] = g.Welo
    lStats["elo"] = g.Lelo

    if i % 2 == 0: 
        feature = np.hstack((wStats, lStats))
        label = 1  
    else:
        feature = np.hstack((lStats, wStats))
        label = 0
        
    if feature.shape[0] == NUM_FEATURES:
        features = np.vstack((features, feature))
        labels = np.vstack((labels, label)) 
print(datetime.now())

np.save("features", features)
np.save("labels", labels)
```

```python
def predictGame(team1, team2, season, daynum, model):
    team1Stats = stats(resultsDf, team1, season, daynum)
    team2Stats = stats(resultsDf, team2, season, daynum)
    
    gameFeatures = np.hstack((team1Stats, team2Stats)).reshape(1, NUM_FEATURES)
    return model.predict_proba(gameFeatures)
```

```python
schoolToId = dict((v,k) for k,v in teams.items())

def getPrediction(predictionsDict, t1, t2):
    t1Id, t2Id = schoolToId[t1], schoolToId[t2]

    if t1Id < t2Id: 
        ts = [t2, t1]
        pred = predictionsDict[(t1Id, t2Id)]
        winningIndex = np.argmax(pred)
        return "Predict {0} will beat {1} with probability {2}".format(ts[winningIndex], ts[~winningIndex], pred[0, winningIndex])      
    else: 
        ts = [t1, t2]
        pred = predictionsDict[(t2Id, t1Id)]
        winningIndex = np.argmax(pred)
        return "Predict {0} will beat {1} with probability {2}".format(ts[winningIndex], ts[~winningIndex], pred[0, winningIndex]) 
    
```
